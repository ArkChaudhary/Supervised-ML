#### One hot encoding

```python
unique_types = ['Grass', 'Fire', 'Water', 'Bug', 'Normal', 'Poison', 'Electric', 'Ground',
 'Fairy', 'Fighting', 'Psychic', 'Rock', 'Ghost', 'Ice', 'Dragon', 'Dark', 'Steel', 'Flying']

for typ in unique_types:
    df[typ] = (df['Type 1'] == typ) | (df['Type 2'] == typ)

df.drop(['Type 1', 'Type 2'], axis=1, inplace=True)
```

```python
# Move 'Legendary' column to the last position
legendary_col = df.pop('Legendary')
df['Legendary'] = legendary_col
```

```python
x = df.loc[:,"Total":"Flying"]
y = df.loc[:,"Legendary"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

# **Models**
predictions

```python
models = {
    "Linear Regression": 0,
    "Logistic Regression": 1,
    "Decision Trees": 2,
    "Decision Trees Gini Index": 3,
    "Decision Trees Entropy": 4,
    "K-NN": 5,
    "Optimized K-NN" : 6,
    "Random Forest" : 7,
    "Naive Bayes" : 8,
    "SVC" : 9,
    "Kernel SVM" : 10
}

predictions = [None] * len(models)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
predictions[models["Linear Regression"]] = model.predict(x_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
predictions[models["Logistic Regression"]] = model.predict(x_test)
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
predictions[models["Decision Trees"]] = clf.predict(x_test)
```

### Gini Index splitting criteria

```python
clf_gini = DecisionTreeClassifier(criterion='gini')
clf_gini.fit(x_train, y_train)
predictions[models["Decision Trees Gini Index"]] = clf_gini.predict(x_test)
```

### Entropy splitting criteria

```python
clf_entropy = DecisionTreeClassifier(criterion='entropy')
clf_entropy.fit(x_train, y_train)
predictions[models["Decision Trees Entropy"]] = clf_entropy.predict(x_test)
```

## K-Nearest Neighbors (K-NN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(x_train, y_train)
predictions[models["K-NN"]] = model.predict(x_test)
```

### Optimized K-NN

```python
# Optimizing parameters

params = {
    "n_neighbors" : [1,2,3,4],
    "algorithm" : ["auto", "ball_tree", "kd_tree", "brute"],
    "leaf_size" : [3,5,7,10]
}
model = KNeighborsClassifier()

grid_search = GridSearchCV(estimator = model, param_grid = params, cv = 5, scoring = "accuracy")
grid_search.fit(x_train, y_train)
best_params_KNN = grid_search.best_params_

model = KNeighborsClassifier(metric_params = best_params_KNN)
model.fit(x_train, y_train)

predictions[models["Optimized K-NN"]] = model.predict(x_test)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 100)
model.fit(x_train, y_train)
predictions[models["Random Forest"]] = model.predict(x_test)
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)

predictions[models["Naive Bayes"]] = model.predict(x_test)
```

## Support Vector Machines

```python
from sklearn.svm import SVC
model = SVC(kernel = "linear")
model.fit(x_train, y_train)

predictions[models["SVC"]] = model.predict(x_test)
```

## Kernel SVM

```python
from sklearn.svm import SVC
model = SVC(kernel="rbf")
model.fit(x_train, y_train)

predictions[models["Kernel SVM"]] = model.predict(x_test)
```

# Evaluation

### Using sklearn library "metrics"
metrics_df

```python
accuracy_scores = []
precision_scores = []
recall_scores = []
mean_squared_errors = []
```

```python
for model_name, index in models.items():
    prediction = predictions[index]

    if prediction is None:
        accuracy_scores.append(None)
        precision_scores.append(None)
        recall_scores.append(None)
        mean_squared_errors.append(None)
        continue

    try:
        accuracy = accuracy_score(y_test, prediction)
    except ValueError:
        accuracy = None

    try:
        precision = precision_score(y_test, prediction)
    except ValueError:
        precision = None

    try:
        recall = recall_score(y_test, prediction)
    except ValueError:
        recall = None

    try:
        mse = mean_squared_error(y_test, prediction)
    except (ValueError, TypeError):
        mse = None

    accuracy_scores.append(accuracy)
    precision_scores.append(precision)
    recall_scores.append(recall)
    mean_squared_errors.append(mse)
```

```python
data = {
    "Model": models.keys(),
    "Accuracy": accuracy_scores,
    "Precision": precision_scores,
    "Recall": recall_scores,
    "MSE": mean_squared_errors,
 }

Result_table = pd.DataFrame(data)
```
